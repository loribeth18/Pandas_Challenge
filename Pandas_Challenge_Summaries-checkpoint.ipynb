```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "C:/Users/lorib/Desktop/Pandas_Challenge-master/schools_complete.csv"
student_data_to_load = "C:/Users/lorib/Desktop/Pandas_Challenge-master/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data_df = pd.read_csv(school_data_to_load)
school_data_df.head()
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete.head()
#school_data_complete.columns
```

```python
#District Summary
#Create a dataframe to hold the above results#
district_summary_df= pd.DataFrame()

#Calculate the total number of schools
district_summary_df.at[0,'total_schools']=len(school_data_complete['school_name'].value_counts().index)

#Calculate the total number of students
district_summary_df.at[0,'total_students']=school_data_complete['Student ID'].count()

#Calculate the total budget
district_summary_df.at[0,'total_budget']=school_data_df['budget'].sum()

#Calculate the average math score 
district_summary_df.at[0,'Avg_Math_Score']=school_data_complete['math_score'].mean()

#Calculate the average reading score
district_summary_df.at[0,'Avg_Reading_Score']=school_data_complete['reading_score'].mean()

#Calculate the percentage of students with a passing math score (70 or greater)
district_summary_df.at[0,'% Passing Math']=len(school_data_complete[school_data_complete['math_score'] >= 70].index) /\
                                                       (school_data_complete['Student ID'].count()) *100

#Calculate the percentage of students with a passing reading score (70 or greater)
district_summary_df.at[0,'% Passing Reading']=len(school_data_complete[school_data_complete['reading_score'] >= 70].index) /\
                                                       (school_data_complete['Student ID'].count())* 100

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
district_summary_df.at[0,'Overall_Score']=((school_data_complete['math_score'].mean() + school_data_complete['reading_score'].mean())/2)

#school_summary_df.head()
district_summary_df.head()

    
```

```python
#School Summary
#Create an overview table that summarizes key metrics about each school, including:
school_summary_df= pd.DataFrame()
```

```python
#School Name
schoolName=  school_data_df['school_name'].values
school_summary_df["school_name"] = schoolName

#School Type
schoolTypes = school_data_df['type'].values
school_summary_df["Types"] = schoolTypes

#Total Students
total_students= school_data_df['size'].values
school_summary_df['Total_Students'] = total_students


#Total School Budget
schoolbudget=school_data_df['budget'].values
school_summary_df["Total__School_Budget"] = schoolbudget

#Per Student Budget
per_student = schoolbudget/total_students
school_summary_df['Per_Student_Budget'] = per_student

print(school_summary_df)
```

```python
school_summary_df=school_summary_df.sort_values(by='school_name').reset_index(drop=True)
print(school_summary_df)
```

```python
schoolgroup=school_data_complete.groupby('school_name')
schoolgroup
```

```python
#Average Math Score
avg_math_score= schoolgroup['math_score'].mean().values
school_summary_df['Average Math Score'] = avg_math_score

#Average Reading Score
avg_rd_score = schoolgroup['reading_score'].mean().values
school_summary_df['Average Reading Score'] = avg_rd_score

print(school_summary_df)
```

```python
passingMath= school_data_complete[(school_data_complete["math_score"] >= 70)]
passingMath=passingMath.sort_values(by='school_name').reset_index(drop=True)
passingMath_group=passingMath.groupby('school_name')
print(passingMath_group)

```

```python
#% Passing Math
pass_math=passingMath_group['math_score'].count().values
pct_pass_math = pass_math/total_students * 100 


school_summary_df['% Passing Math'] = pct_pass_math
print(school_summary_df)
```

```python
#Reading Group by
passingReading = school_data_complete[(school_data_complete["reading_score"] >= 70)]
passingReading=passingReading.sort_values(by='school_name').reset_index(drop=True)
passingReading_group=passingReading.groupby('school_name')
pass_reading=passingReading_group['reading_score'].count().values
print(pass_reading)
```

```python
#% Passing Reading
pass_reading=passingReading_group['reading_score'].count().values
pct_pass_reading = pass_reading/total_students * 100 

school_summary_df['% Passing Reading'] = pct_pass_reading

print(school_summary_df)
```

```python
#Overall Passing Rate (Average of the above two)
school_summary_df['Overall_School_Score']=((pct_pass_math + pct_pass_reading )/2)
print(school_summary_df)
```

```python
#create df for school spending based in schoolsummarydf
school_spending_df=school_summary_df[['Per_Student_Budget', 'Average Math Score', 'Average Reading Score',
       '% Passing Math', '% Passing Reading', 'Overall_School_Score']]
school_spending_df
```

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
# Create labels for these bins
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# Slice the data and place it into bins
pd.cut(school_spending_df['Per_Student_Budget'],spending_bins, labels=group_names)  
```

```python
# Place the data series into a new column inside of the DataFrame
school_spending_df["Spending Ranges per student"]=pd.cut(school_spending_df['Per_Student_Budget'],spending_bins, labels=group_names)   
school_spending_df.head() 
```

```python
# Create a GroupBy object based upon "View Group"
spent_groups_df=school_spending_df.groupby("Spending Ranges per student")

# Find how many rows fall into each bin
print(spent_groups_df["Overall_School_Score"].count())

# Get the average of each column within the GroupBy object
spent_groups_df[['Average Math Score', 'Average Reading Score',
       '% Passing Math', '% Passing Reading', 'Overall_School_Score']].mean()
```
